```python
import rawpy
import PIL
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
raw = rawpy.imread('/home/anton/DSC01660.ARW')
```

```python
#reading raw image
mosaic = raw.raw_image

img_visualize = np.float32(mosaic) 

#   

plt.imshow(img_visualize, cmap='gray')
plt.axis('off')
plt.title('the raw image')
```

```python
#linearization
black = mosaic.min() #     -
saturation = mosaic.max()

uint14_max = 2**14 - 1

mosaic -= black # black subtraction
mosaic *= int(uint14_max/(saturation - black))
mosaic = np.clip(mosaic,0,uint14_max) # 

mosaic = np.float64(mosaic)

#   
mg_visualize = np.float32(mosaic)
plt.imshow(img_visualize, cmap='gray')
plt.axis('off')
plt.title('linearized image')
```

```python
# white-balancing
wb = np.asarray(raw.camera_whitebalance) #   

wb /= wb[1]                # scale green to 1
mosaic[0::2, 0::2] *= wb[0]     # scale reds
mosaic[1::2, 1::2] *= wb[2]     # scale blues

mosaic = np.clip(mosaic,0,uint14_max)
img_visualize = np.float32(mosaic)
plt.imshow(img_visualize, cmap='gray')
plt.axis('off')
plt.title('whitebalanced image')
```

```python
def demosaic(m):
    r = m[0::2, 0::2]
    g = np.clip(m[0::2, 1::2]//2 + m[1::2, 0::2]//2,
                0, 2 ** 16 - 1)
    b = m[1::2, 1::2]
    return np.dstack([r, g, b])

mosaic *= 2**2    
img = demosaic(mosaic)          #  rgb 
img_visualize = np.float32(img) / 2**16
plt.imshow(img_visualize)
plt.axis('off')
plt.title('demosaiced image')

```

```python
#color space convertion
cam2xyz = raw.rgb_xyz_matrix[0:3, :]
xyz2cam = np.linalg.inv(cam2xyz)
#     
srgb2xyz = np.array([[0.4124564,0.3575761,0.1804375],[0.2126729,0.7151522,0.0721750],[0.0193339,0.1191920,0.9503041]])
srgb2cam = srgb2xyz * xyz2cam

row_sums = srgb2cam.sum(axis=1) #   
new_matrix = srgb2cam / row_sums[:, np.newaxis]

cam2srgb = np.linalg.inv(new_matrix) 

cam2srgb = np.round(cam2srgb*255).astype(np.int16) #   
img = img // 2**8          
shape = img.shape # 

pixels = img.reshape(-1, 3).T     #   3N
pixels = cam2srgb.dot(pixels)//255

img = pixels.T.reshape(shape)
img = np.clip(img, 0, 255).astype(np.uint8)

plt.imshow(img)
plt.axis('off')
plt.title('image in linear sRGB color space')
```
